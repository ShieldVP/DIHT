Выполнив это задание, вы увидели на практике, чем поиск минимума функции отличается от глобальной оптимизации, и когда может быть полезно применить вместо градиентного метода оптимизации метод, не использующий градиент. Кроме того, вы попрактиковались в использовании библиотеки SciPy для решения оптимизационных задач, и теперь знаете, насколько это просто и удобно.

```python
# Задаём функцию f(x)
from math import sin, exp
import numpy as np
def f(x):
    return sin(x[0] / 5.0) * exp(x[0] / 10.0) + 5 * exp(-x[0] / 2.0)
```

```python
# Реализуем минимизацию через BFGS
from scipy.optimize import minimize
def opt(f, x):
    res = minimize(f, (x), method='BFGS')
    return res.fun
print(opt(f, 2), opt(f, 30))
```

```python
# Реализуем минимизацию через дифференциальную эволюцию на [1, 30]
from scipy.optimize import differential_evolution
def dif(f):
    res = differential_evolution(f, [(1, 30)])
    return res.fun
print(dif(f))
```

```python
# Задаём функцию h(x)
def h(x):
    return int(f(x))
```

```python
print(opt(h, 30), dif(h)) # Сравниваем минимумы h(x), полученные двумя методами
```
